```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
```

```python
net = pd.read_csv('dataset/Internet_demographic.csv')
```

```python
net
```

```python
data = {
    "Country": ["Country A", "Country B", "Country C", "Country D", "Country E",
                "Country F", "Country G", "Country H", "Country I", "Country J"],
    "Birth Rate": [35, 42, 27, 50, 38, 18, 45, 33, 29, 41],
    "Internet Usage (%)": [80, 75, 90, 60, 85, 95, 50, 70, 88, 40],
    "Income Group": ["High", "Middle", "Low", "Low", "Middle",
                     "High", "Low", "Middle", "High", "Low"]
}
```

```python
df = pd.DataFrame(data)
```

<h1>Distribution of Birth Rates using Histogram</h1>

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Birth Rate"], bins=5, kde=True)
plt.xlabel("Birth Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Birth Rates")
plt.show()
```

<h1> Top 5 Countries with Highest Internet Usage</h1>

```python
top_5 = df.nlargest(5, "Internet Usage (%)")
plt.figure(figsize=(8, 5))
sns.barplot(x="Country", y="Internet Usage (%)", data=top_5)
plt.xlabel("Country")
plt.ylabel("Internet Usage (%)")
plt.title("Top 5 Countries with Highest Internet Usage")
plt.xticks(rotation=45)
plt.show()
```

<h1> Count of Countries by Income Group</h1>

```python
plt.figure(figsize=(6, 5))
sns.countplot(x=df["Income Group"], order=["Low", "Middle", "High"])
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.title("Number of Countries by Income Group")
plt.show()
```

<h1> Box Plot of Internet Usage</h1>

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y=df["Internet Usage (%)"])
plt.ylabel("Internet Usage (%)")
plt.title("Box Plot of Internet Usage")
plt.show()
```

<h1>Countries with No Internet Access</h1>

```python
no_internet = df[df["Internet Usage (%)"] == 0]
print("Countries with No Internet Access:", no_internet)
```

<h1>Average Birth Rate</h1>

```python
avg_birth_rate = net["Birth rate"].mean()
print(f"Average Birth Rate: {avg_birth_rate:.2f}")
```

<h1> Scatter Plot: Birth Rate vs. Internet Usage</h1>

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Birth Rate", y="Internet Usage (%)", data=df)
plt.xlabel("Birth Rate")
plt.ylabel("Internet Usage (%)")
plt.title("Scatter Plot: Birth Rate vs. Internet Usage")
plt.show()
```

<h1>Countries with Birth Rate Above 40</h1>

```python
high_birth_rate = df[df["Birth Rate"] > 40]
print("Countries with Birth Rate Above 40:\n", high_birth_rate)
```

<h1>Pie Chart of Income Groups</h1>

```python
plt.figure(figsize=(6, 6))
df["Income Group"].value_counts().plot.pie(autopct='%1.1f%%', colors=["red", "blue", "green"])
plt.title("Proportion of Countries by Income Group")
plt.ylabel("")  # Hide y-label
plt.show()
```

<h1>Correlation Between Birth Rate and Internet Usage</h1>

```python
correlation = df["Birth Rate"].corr(df["Internet Usage (%)"])
print(f"Correlation between Birth Rate and Internet Usage: {correlation:.2f}")
```
